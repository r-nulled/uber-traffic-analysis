```python
from fbprophet import Prophet
```

```python
baseList = picked["osm_end_node_id"].value_counts()

locList = list(baseList.index)
locList
```

```python
llocL = locList[0:45]
llocL
```

```python
storageMet = {}
storageAvgRMSE = {}
strAvgMae = {}
strAvgMape = {}
for loc in llocL:
    isPlace = picked["osm_end_node_id"] == loc
    restruc = picked[isPlace]
    reb = restruc.drop(["osm_way_id", "osm_start_node_id", "osm_end_node_id"], axis = 1)
    
    bal = reb.rename(columns = {"Date" : "ds", "speed_mph_mean" : "y"})
    
    train_test = np.array_split(bal, 2)
    train = train_test[0]
    
    model = Prophet()
    
    
    model = model.add_regressor('Temperature')
    model = model.add_regressor('Humidity')
    model = model.add_regressor('WindSpeed')
    model = model.add_regressor('pedestrians')
    model = model.add_regressor('to_manhattan')
    model = model.add_regressor('NYIS demand megawatthours')
    model = model.add_regressor('NYIS demand forecast megawatthours')
    model = model.add_regressor('tot')
    
    
    g = model.fit(train)
    
    
    test = train_test[1]
    test_vars = test.drop("y", axis = 1)    
    forecast = model.predict(test_vars)
    
    
    df_cv = forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
    df_cv = pd.concat([df_cv, test["y"].reset_index()], axis = 1)
    df_cv.drop("index", axis = 1, inplace = True)
    df_cv = df_cv.groupby("ds").mean()
    df_cv = df_cv.reset_index()
    df_cv["cutoff"] = pd.Timestamp("2020-02-15")
    
    metrics = fbprophet.diagnostics.performance_metrics(df_cv)
    
    storageMet[loc] = metrics
    
    p = list(metrics['rmse'])
    v = min(p)
    storageAvgRMSE[loc] = v
    
    l = list(metrics['mae'])
    m = min(l)
    strAvgMae[loc] = m
    
    j = list(metrics['mape'])
    o = min(j)
    strAvgMape[loc] = o
    
```
